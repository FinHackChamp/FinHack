```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
user_name = 'David Mendoza'
```

```python
df = pd.read_csv('transaction.csv')
df
```

```python
user_age = list(df.loc[df['name'] == user_name, 'age'])[0]
```

```python
#generate same age dataframe
dfSameAge = df.loc[(df['age'] < 50 )]
dfSameAge = dfSameAge.loc[dfSameAge['age'] >= 40]
dfSameAge
```

```python
# prepare for the array
companies = pd.read_csv('label.csv')
company_names= list(companies.name)
company_names
```

```python
number_of_people = int(dfSameAge.shape[0] / 40)

allList = []
for i in range(number_of_people):
    currentList = []
    for j in range(40):
        row= dfSameAge.iloc[i*40+j]
        companyName = row['company name']
        
        companyIndex = company_names.index(companyName)
        oneHot = np.zeros(len(company_names))
        oneHot[companyIndex] = 1
        currentList.append(oneHot)
    allList.append(currentList)
np.array(allList).shape
```

```python
np.save('ontHotEncoded.npy', allList)
```
