```python
import pandas as pd
import numpy as np
import random
```

```python
names = pd.read_csv('transaction.csv')['name']
names = pd.DataFrame(list(set(names)), columns=['name'])
transaction = pd.read_csv('transaction.csv')
```

```python

salaryList= []
for i in range(names.shape[0]):
    if i %1000 == 0: print(i)
    name = names.iloc[i]['name']
    total = sum(transaction.loc[transaction['name'] == name, 'amount'])
    percentage = random.randrange(-10,10000)
    
    total += total * (percentage/100.0)
    
#     print(total, percentage)
    salaryList.append(int(total))
    
    
names['salary'] = salaryList
names
```

```python
transaction = pd.read_csv('transaction.csv')
transaction
```

```python
names = names[['name', 'salary']]
np.mean(names.salary)
```

```python
transaction = transaction.join(names.set_index('name'), on=['name'])
```

```python
transaction.to_csv('transaction.csv', index = False)
```
